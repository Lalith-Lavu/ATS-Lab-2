```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller

# ----------------------------------------
# 1. Create Stationary Series C
# ----------------------------------------
np.random.seed(42)
n = 200
series_C = np.random.normal(loc=5, scale=1, size=n)

# ----------------------------------------
# 2. Plot the Autocorrelation Function (ACF)
# ----------------------------------------
plt.figure(figsize=(10,5))
plot_acf(series_C, lags=30)
plt.title("ACF of Stationary Series C")
plt.show()

# ----------------------------------------
# 3. ADF Test for Stationarity
# ----------------------------------------
result = adfuller(series_C)

print("ADF Test Results")
print("-" * 40)
print(f"ADF Statistic      : {result[0]}")
print(f"p-value            : {result[1]}")
print("Critical Values:")
for key, value in result[4].items():
    print(f"   {key} : {value}")
print("-" * 40)

if result[1] < 0.05:
    print("Conclusion: p-value < 0.05 → Reject null hypothesis.")
    print("The series IS stationary.")
else:
    print("Conclusion: p-value ≥ 0.05 → Fail to reject null hypothesis.")
    print("The series is NOT stationary.")

# ----------------------------------------
# 4. Summary of Results (Printed)
# ----------------------------------------
summary = """
SUMMARY:

• The ACF plot of stationary data drops very quickly because stationary data
  has no long-term dependence. Each value is mostly independent of distant past values.

• Trending/seasonal data has slowly decaying ACF because:
    - Trend introduces long-term correlation.
    - Seasonality creates repeating correlations across cycles.

• ADF (Augmented Dickey-Fuller) Test confirms stationarity when:
    - ADF Statistic is negative and large in magnitude.
    - p-value < 0.05 (rejects null hypothesis of non-stationarity)

• Based on the ADF test above, Series C is stationary.
"""

print(summary)
```
